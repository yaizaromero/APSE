```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import pickle

# Seleccionar caracterÃ­sticas relevantes y la variable objetivo para el modelo de tiempo de entrega de los paquetes
X_camiones = entregas[['eventTime']].values
y_camiones = entregas[['delta']].values

# Dividir los datos en conjuntos de entrenamiento y prueba para el modelo de tiempo de entrega de los paquetes
X_train_camiones, X_test_camiones, y_train_camiones, y_test_camiones = train_test_split(X_camiones, y_camiones, test_size = 0.25, train_size = 0.75, random_state=42)

# Entrenar el modelo de regresiÃ³n lineal para el tiempo de entrega de los paquetes
modelo_camiones = LinearRegression()
modelo_camiones.fit(X_train_camiones, y_train_camiones)

# Evaluar el modelo de tiempo de entrega de los paquetes
y_pred_camiones = modelo_camiones.predict(X_test_camiones)
mse_camiones = mean_squared_error(y_test_camiones, y_pred_camiones)
print("Error cuadrÃ¡tico medio para tiempo de entrega de paquetes:", mse_camiones)

# Guardar el modelo entrenado para el tiempo de entrega de los paquetes
with open('modelo_tiempo_entrega_paquetes.pkl', 'wb') as f_camiones:
    pickle.dump(modelo_camiones, f_camiones)
```
